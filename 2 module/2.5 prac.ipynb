```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

brfss = pd.read_hdf('brfss.hdf5')
height = brfss['HTM4']
weight = brfss['WTKG3']

plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.scatter(height, weight, marker='o')
plt.xlabel('Рост в см')
plt.ylabel('Вес в кг')
plt.title('отношение роста к весу')
plt.tight_layout()
noise = np.random.normal(0, 2, size=len(brfss))
height_noise = height + noise
weight_noise = weight + noise
plt.subplot(1, 2, 2)
plt.scatter(height_noise, weight_noise, marker='o', alpha=0.02, s=1)
plt.xlim([140, 200])
plt.ylim([0, 160])
plt.xlabel('Рост в см')
plt.ylabel('Вес в кг')
plt.title('отношение роста к весу')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

data = brfss.dropna(subset=['AGE', 'WTKG3'])
sns.boxplot(x='AGE', y='WTKG3', data=data, whis=10)
```

```python
import seaborn as sns
data = brfss.dropna(subset=['AGE', 'WTKG3', 'HTM4','SEX'])
data = data.drop_duplicates()
plt.subplot(1, 2, 1)
sns.violinplot(x='AGE', y='WTKG3', data=data, inner=None)
plt.xlabel('Возраст')
plt.ylabel('Вес')
plt.title('Возраст вес')
plt.tight_layout()
plt.subplot(1, 2, 2)
sns.boxplot(x='AGE', y='WTKG3', data=data, whis=10)
plt.xlabel('Возраст')
plt.ylabel('Вес')
plt.title('Возраст вес')
plt.tight_layout()
plt.show()
```

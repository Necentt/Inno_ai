```python
import tensorflow as tf
from keras.layers import Dense
# Создаем данные для обучения
X_train = [[1.0, 2.0], [2.0, 3.0], [3.0, 4.0], [4.0, 5.0]]
y_train = [[3.0], [5.0], [7.0], [9.0]]
```

```python
# Создаем модель нейронной сети
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, input_shape=(2,), activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
# Компилируем модель
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучаем модель
model.fit(X_train, y_train, epochs=200, verbose=False)
```

```python
# Предсказываем результат сложения двух чисел
result = model.predict([[5.0, 6.0]])
print(result)
```

```python
import tensorflow as tf

# Создаем данные для обучения
X_train = [[1.0, 2.0], [2.0, 3.0], [3.0, 4.0], [4.0, 5.0]]
y_train = [[3.0], [5.0], [7.0], [9.0]]
```

```python
# Создаем модель нейронной сети с одним слоем и одним нейроном
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2,))
])
# Компилируем модель
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python

# Обучаем модель
model.fit(X_train, y_train, epochs=1000)
```

```python
# Предсказываем результат сложения двух чисел
result = model.predict([[-5.0, 16.0]])
print(result)
```
